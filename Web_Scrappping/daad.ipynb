```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
soup.prettify()
```

```python
box = soup.find_all('li', class_ = 'qa-db-result-item js-db-result-item u-focus-outline--none')
course_name = []
uni_name = []
for i in box:
  course_name.append(i.find('span',class_ = 'js-dynamic-content u-display-block u-font-light u-size-24 mb-24 u-size-32@lg').text.strip())
  uni_name.append(i.find('span',class_ = 'js-dynamic-content u-display-block u-font-regular u-size-18 mb-8').text.strip())
```

```python
data = pd.DataFrame({'Uni_name' : uni_name, 'Course_Name':course_name})
data
```

```python
all_links = []
for j in range(1,3):
    url = f'https://www.daad.de/en/studying-in-germany/universities/all-degree-programmes/?hec-degreeProgrammeType=w&hec-subjectGroup=56&hec-location=15&hec-p={j}'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36"
        }
    webpage = requests.get(url,headers=headers).text
    
    soup = BeautifulSoup(webpage, 'html.parser')
    soup.prettify()
    
    box = soup.find_all('li', class_ = 'qa-db-result-item js-db-result-item u-focus-outline--none')
    
    links = []
    for i in box:
        links.append(i.find('a', href = True))
    for k in links:    
        all_links.append(k['href'])
```
