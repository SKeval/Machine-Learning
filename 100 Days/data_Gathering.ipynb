# With CSV


```python
# import data from url 

import requests
import pandas as pd
from io import StringIO

url = "" # paste URL into this
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

```python
# Sep Parameter , to define Seperater
pd.read_csv(sep='\t')
```

```python
# Index_col , to make column index
pd.read_csv(index_col= 'Column_name')
```

```python
#Name Parameter, to make header
pd.read_csv(names=['names of columns'])
```

```python
# Header Parameter , from where you want to start rows
pd.read_csv(header=1)
```

```python
# Use_col parameter, Which columns you want to use
pd.read_csv(usecols=['names of columns'])
```

```python
# Squeeze PArameter , to make dataframe a series
pd.read_csv(usecols=['names of columns'], squeeze = True)
```

```python
# Skip rows/ nrows parameter,  to skip this rows and Nraws = to add this number of rows
pd.read_csv(skiprows=['number of columns (0, 1)'])
```

```python
# Encoding parameter, change encoding when you will get UnicodeDecodeError
pd.read_csv(encoding='')
```

```python
# Skip bad lines , if all lines have no same values skip with different values
pd.read_csv(error_bad_lines = False)
```

```python
# Dtypes parameter, to change data type
pd.read_csv(dtype={'column name':int})
```

```python
#Handling Dates, to change date type to date from object
pd.read_csv(parse_dates=['column name'])
```

```python
# Converters, to convert to short name or other names
def rename(name):
    if name == 'RCB':
        return 'rcb'
    else:
        return 'RCB'
    
rename('RCB')

pd.read_csv(converters={'column-name' : rename})
```

```python
# Na_Values Parameter, consider other value as NA

pd.read_csv(na_values=['value name'])
```

```python
# Loading Huge data as Chunks, devide big data into chunks, it will devide data in small chunks
df = pd.read_csv(chunksize=5000)
```
