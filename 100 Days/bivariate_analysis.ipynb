```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df_tip = sns.load_dataset('tips')
df_tip.head()
```

## ScatterPlot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex',style='smoker', data=df_tip, size='size')
```

## Bar Plot (Numerical - Categorical)

```python
df_titanic = sns.load_dataset('titanic')
df_titanic.head()
```

```python
sns.barplot(x='pclass', y='age',data=df_titanic, hue='survived')    # X = Categorical, y = Numerical
```

## Box Plot (Numerical - Categorical)

```python
sns.boxplot(x='sex', y='age',data=df_titanic)
```

## Distplot (Numerical - Categorical)

```python
sns.displot(df_titanic[df_titanic['survived']==0]['age'], kind='kde')
sns.displot(df_titanic[df_titanic['survived']==1]['age'], kind='kde')
```

## HeatMap (Categorical - Categorical)

```python

sns.heatmap(pd.crosstab(df_titanic['pclass'],df_titanic['survived']), annot=True)
```

```python
pd.crosstab(df_titanic['pclass'],df_titanic['survived'])
```

## ClusterMap (Categorical - Categorical)

```python
import scipy

sns.clustermap(pd.crosstab(df_titanic['parch'],df_titanic['survived']))
```
