```python
data = pd.read_csv('50_Startups.csv')
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
#multiple Linear Regression 

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[3])], remainder='passthrough')
x = np.array(ct.fit_transform(x))

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

```python
reg.fit(X_train, Y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
df = pd.DataFrame({'Real Value': Y_test,'Predicted Value':y_pred})
```

```python
df
```

```python
#Root Mean Square Error = to find how good is the accuracy of our model, 
# Mean-squared error useful to evaluate model while modeling a continuous output variable.

from sklearn import metrics
print(np.sqrt(metrics.mean_absolute_error(Y_test,y_pred)))
```

```python
import seaborn as sns
sns.displot(y_pred-Y_test, kind='kde')
```
